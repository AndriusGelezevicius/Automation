```python
# Atidaryk https://www.python.org ir surask pirmą h1 elementą:
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get("https://www.python.org")

heading = driver.find_element(By.TAG_NAME, "h1")
print(heading.text)

driver.quit()
```

```python
# teksto įvedimas ir mygtuko paspaudimas

from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get("https://www.selenium.dev/selenium/web/web-form.html")

text_box = driver.find_element(By.NAME, "my-text")
submit_button = driver.find_element(By.CSS_SELECTOR, "button")

text_box.send_keys("Andriud mokosi Selenium")
submit_button.click()

message = driver.find_element(By.ID, "message")
print(message.text)

driver.quit()
```

```python
# Tau reikia parašyti Selenium testą, kuris užpildo formą ir patikrina gautą rezultatą.
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Chrome()
driver.get("https://www.selenium.dev/selenium/web/web-form.html")

text_input = driver.find_element(By.ID, "my-text-id")
submit_button = driver.find_element(By.CSS_SELECTOR, "button")

text_input.send_keys("My first automated test")
submit_button.click()

message = driver.find_element(By.ID, "message")
print(message.text)

assert message.text == "Received!"
driver.quit()
```
