```python
# Series – vienos dimensijos duomenų masyvas (panašus į sąrašą).

import pandas as pd

data = [10, 20, 30, 40]
s = pd.Series(data)
print(s)

data2 = {"Alice": 80, "Bob": 75, "Andrew": 10}
s1 = pd.Series(data2)
print(s1[s1>70])
print(s1.mean())
```

```python
# Duomenu rusiavimas
import pandas as pd

df = pd.DataFrame({
    "Name": ["Anna", "Jonas", "Tomas", "Laura", "Petras", "Asta"],
    "Age": [28, 34, 22, 45, 34, 28],
    "City": ["Vilnius", "Kaunas", "Klaipėda", "Šiauliai", "Kaunas", "Vilnius"]
})
df.sort_values(by="Age", inplace=True)
print(df)

s = df.groupby("Name")["Age"].mean()
print(s)
```
