```python
import pandas as pd

df = pd.DataFrame({
    "Name": ["Anna", "jonas", " Jonas", " Laura "],
    "Age": [28, 34, 34, 45]
})

f =df.duplicated()
print(f)
s = df.drop_duplicates()
print(s)

df["Age"] = df["Age"].astype(str)
df["Name"] = df["Name"].str.strip()
df["Name"] = df["Name"].str.lower()
df["Name"] = df["Name"].str.title()
df["Name"] = df["Name"].str.replace("Jonas", "Andrius")
print(df)
```

```python
import pandas as pd
df = pd.DataFrame({
    "Name": ["Anna", "Jonas", "Jonas", "Laura"],
    "Age": ["28", "34", "34", "45"]
})
df["Age"] = df["Age"].astype(int)
vid = df["Age"].mean()
print(vid)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Įmonė": ["Įmonė A", "Įmonė B", "Įmonė C"],
    "Sektorius": ["Technologijos", "Finansai", "Sveikata"]
})
sector_dict = {
    "Technologijos": "Tech",
    "Finansai": "Fin",
    "Sveikata": "Health"
}
df["Sektoriaus satrumpa"] = df["Sektorius"].map(sector_dict)
print(df)

#####

df = pd.DataFrame({
    "Darbuotojas": ["Jonas", "Laura", "Tomas", "Asta"],
    "Statusas": ["Pilnas etatas", "Pusė etato", "Pilnas etatas", "Pusė etato"]
})

status_dict = {
    "Pilnas etatas": 1,
    "Pusė etato": 0.5
}

df["Etatas"] = df["Statusas"].map(status_dict)
print(df)

#####

df = pd.DataFrame({
    "Vardas": ["Jonas", "Laura", "Tomas", "Asta"],
    "Gimimo metai": [1990, 1985, 1995, 1992]
})

df["Amzius"] = df["Gimimo metai"].map(lambda x: 2025 - x)
print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Vardas": ["Jonas", "Laura", "Tomas"],
    "Pavardė": ["Povilas", "Petrauskaitė", "Jonaitis"]
})

df["Pilnas vardas"] = df["Vardas"] + " " + df["Pavardė"]
df["Pilnas_vardas"] = df["Vardas"].str.cat(df["Pavardė"], sep=" ")
print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    "Vardas": ["Jonas", "Laura", "Tomas"],
    "Amžius": [25, 30, 22],
    "Miesto tipas": ["Miestas", "Miestas", "Kaimas"]
})

# Sukuriame naują stulpelį, kuris apjungia vardą, amžių ir miesto tipą pagal amžiaus sąlygą
df["Vardas_amžius_miestas"] = df.apply(lambda row: f"{row['Vardas']} ({row['Amžius']}), {row['Miesto tipas']}" if row['Amžius'] >= 30 else f"{row['Vardas']} ({row['Amžius']}), jaunimas", axis=1)

print(df)
```
