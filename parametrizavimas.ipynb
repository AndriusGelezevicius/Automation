```python
 # Failų struktūra

projektas/
├── kalkuliatorius.py
└── test_kalkuliatorius.py

# kalkuliatorius.py

def suma(a, b):
    return a + b

def atimtis(a, b):
    return a - b

def sandauga(a, b):
    return a * b

def dalyba(a, b):
    if b == 0:
        return "Dalyba iš nulio negalima"
    return a / b



# test_kalkuliatorius.py

import pytest
from kalkuliatorius import suma, atimtis, sandauga, dalyba

# --- Testas funkcijai `suma`
@pytest.mark.parametrize("a, b, rezultatas", [
    (2, 3, 5),
    (0, 0, 0),
    (-1, 1, 0),
    (-5, -5, -10),
])
def test_suma(a, b, rezultatas):
    assert suma(a, b) == rezultatas


# --- Testas funkcijai `atimtis`
@pytest.mark.parametrize("a, b, rezultatas", [
    (5, 3, 2),
    (0, 0, 0),
    (-2, -3, 1),
])
def test_atimtis(a, b, rezultatas):
    assert atimtis(a, b) == rezultatas


# --- Testas funkcijai `sandauga`
@pytest.mark.parametrize("a, b, rezultatas", [
    (2, 3, 6),
    (0, 10, 0),
    (-2, 3, -6),
])
def test_sandauga(a, b, rezultatas):
    assert sandauga(a, b) == rezultatas


# --- Testas funkcijai `dalyba`
@pytest.mark.parametrize("a, b, rezultatas", [
    (10, 2, 5),
    (3, 2, 1.5),
    (5, 0, "Dalyba iš nulio negalima"),
])
def test_dalyba(a, b, rezultatas):
    assert dalyba(a, b) == rezultatas
```
