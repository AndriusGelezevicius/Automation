### Testų parametrizavimas

```python
# Kas yra parametrizavimas?
# - Testų parametrizavimas leidžia:
# - paleisti tą patį testą daug kartų su skirtingais įėjimo duomenimis.
# - išvengti pasikartojančio kodo.

# Paprastas pavyzdys:
import pytest

@pytest.mark.parametrize("a, b, expected", [
    (2, 3, 5),
    (1, 1, 2),
    (5, 5, 10),
])
def test_addition(a, b, expected):
    assert a + b == expected
# Rezultatas:
# Testas bus paleistas 3 kartus:
# a=2, b=3 → expected=5
# a=1, b=1 → expected=2
# a=5, b=5 → expected=10

# Sintaksė:
@pytest.mark.parametrize("param1, param2", [
    (val1, val2),
    (val3, val4),
])
def test_something(param1, param2):
    ...
```
